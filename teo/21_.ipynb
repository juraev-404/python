```python
import scipy.stats as sts
import numpy as np
import math
```

```python
aplha = 0.05
mu0 = 35
xi = np.array([34.8, 34.9, 35, 35.1, 35.3])
ni = np.array([2, 3, 4, 6, 5])
```

```python
n = sum(ni)
x_sr = sum(xi * ni) / n
s2 = sum(ni * (xi - x_sr) ** 2) / (n - 1)
s_ = math.sqrt(s2)
se = s_ / math.sqrt(n)
t_obs = (x_sr - mu0) / se
t_crit = sts.t.ppf(1 - aplha, df=n - 1)
print("t наблюдаемое:", t_obs)
print("t критическое:", t_crit)
```
